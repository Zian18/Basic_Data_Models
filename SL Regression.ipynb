```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('satgpa.csv')
```

```python
data
```

```python
data.describe()
```

```python
y = data['hs_gpa']
x1 = data['sat_sum']
```

```python
plt.scatter(x1,y)
plt.xlabel('sat_sum',fontsize=20)
plt.ylabel('hs_gpa',fontsize=20)
plt.show()
```

```python
x=sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat= 0.0163*x1 + 1.5106
fig = plt.plot(x1,yhat, lw=4, c='orange', label = 'regressionl line')
plt.xlabel('sat_sum',fontsize=20)
plt.ylabel('hs_gpa',fontsize=20)
plt.show()
```
