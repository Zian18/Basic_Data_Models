```python
raw_data = pd.read_csv('1.03. Dummies.csv')
```

```python
data1 = raw_data.copy()
```

```python
data1['Attendance']=data1['Attendance'].map({'Yes':1, 'No':0})
```

```python
data1
```

```python
data1.describe()
```

```python
y = data1['GPA']
x1 = data1[['SAT', 'Attendance']]
```

```python
x= sm.add_constant(x1)
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```
